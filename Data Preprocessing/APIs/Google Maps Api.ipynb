A. Use the [*Nearby Seach*](https://developers.google.com/places/web-service/search) Method to find all the restaurant near to the UPY (20.988459, -89.736768) around 2km. In order to fulfill this item you need to:
* Create a *search_nearby_places* function with the appropiate parameters. It must return a JSON with the results. Places must be ranked by distance.
* Create a table with the following columns: Name, Place Id, Rating, Place Types, Total of User Ratings.

```python
# Importing the libraries 
import pandas as pd
import json
import requests
```

```python
# Variable key that is gonna be use in every one of the fuctions
key = ''
```

```python
#Definate the function

def search_nearby_places(key, location, radius, rankby=None, typee=None):
    """
    This functions is use to find places near to a specific positions
    INPUTS:
    Key: Key get from google Cloud
    Location: Latitude and Longitude of the places tu use as the point to reference 
    Radius: Amount of meters where the functions is gonna find (in meters)
    rankby: Order in the results are gonna be return 
    OUTPUT: 
    A json file in a dictionary as a general list definate
    
    """
    #EndPoint
    url="https://maps.googleapis.com/maps/api/place/nearbysearch/json"
    #Parameters taken from the functions
    parameters={
        'key': key,
        'location':location, 
        'radius':radius,
        'rankby':rankby,
        'type':typee
      }
    #Resquest
    res=requests.get(url,params=parameters)
    #Data Parsing
    result=json.loads(res.content)
    return result
```

B. Create a function to Find the Place Id given Place Name (String) and Location (Lat/Lont Tuple).

```python
def place_id(key, name, text, rad, lat,long):#Defining the functions
    """
    This functions return the ID of a place according to the following parameters
    INPUTS:
    Key: Key get from google Cloud
    Name: Name of the place (string)
    Text: Textquery or even can be a number, but in this case we use the definite textquery
    Location: Latitude and Longitude of the places tu use as the point to reference 
    Radius: Amount of meters where the functions is gonna find (in meters)
    OUTPUT:
    Json with only the id of the place
    """
    parameters={
        'key':key,
        'input':name,
        'inputtype':text,
        'locationbias':'circle:'+str(rad)+'@'+str(lat)+','+str(long)
    }
    #EndPoint
    url = "https://maps.googleapis.com/maps/api/place/findplacefromtext/json"
    #Request
    res = requests.get(url, params=parameters)
    #Data Parsing
    result = json.loads((res.content))
    result = result['candidates'][0]['place_id']
    return result
```

E. Find similar places to Los Trompos Circuito. Rank them by its "Prominence". In order to fulfill this item you need to:
* Create a *find_similar* function with the appropiate parameters. It must return a JSON with the results.
* Create a table with the following columns: Place Name, Plaece Id, Rating, Total of Use Ratings.

```python
#Definate the function

def search(key,name, ipt ='textquery'): #Ination of the function
    """
    INPUTS:
    key: Get from google cloud
    name: Name of the place to find
    ipt: Input of textquery definite in this case
    fields: Specific data to return
    OUTPUT:
    A json with the specific fields
    """
    
    parameters= { 
        'key': key,
        'input': name,
        'inputtype': ipt,
        'fields': 'geometry/location,type'
    }
    #Endpoint
    url="https://maps.googleapis.com/maps/api/place/findplacefromtext/json"
    #Request
    res=requests.get(url,params=parameters)
    #Data Parsing 
    result=json.loads(res.content)
    return result
```

```python
def find_similar(key, name, radius, rankby='prominence'):
    """
    This functions return similar place according a place and locations
    INPUTS:
    key: Get from google chrome
    name: Name of the place to find (this is send to the search functions)
    radius: Distance to find around of the location
    ranby: Results return accorging to the prominece
    """
    #Calling the function search and saving in a json file
    places = search(key, name, ipt='textquery')
    #Gettign the latitude of the place 
    lat = str(places['candidates'][0]['geometry']['location']['lat'])
    #Getting the longitude of the place
    long = str(places['candidates'][0]['geometry']['location']['lng'])
    #Returning the kind of place find
    types= str(places['candidates'][0]['types'][0])
    #Endpoint
    url = 'https://maps.googleapis.com/maps/api/place/nearbysearch/json'
    #Parameters used
    parameters2= {
        'key':key,
        'location':lat+','+long,
        'type': types,
        'radius': radius,
        'rankby': rankby
    }
    #Request
    res = requests.get(url, params=parameters2)
    #Data Parsing 
    result = json.loads(res.content)
    return result

                                   
```

## Finding similar places

```python
#Intermediate Functions:
def first(name, input_type=None):
    """
    This is a intermediate functions
    INPUT:
    name: Name of the place to find
    Input: A predeminate parameters
    Output: : Json file with candidates
    """
    result = gmaps.find_place(name, input_type)
    return result
```

```python
#Second Intermediate Function
def second():
    """This return the location of the place found before
    INPUT: None
    OUPUT: Locations as json dictionary
    """
    result = first(name='Los Trompos Circuito',input_type='textquery')
    result = result['candidates'][0]['place_id']
    new_result = gmaps.place(result)
    return new_result
```

```python
def find_similargp(rank='prominence'):
    result = second()
    lat=str(result['result']['geometry']['location']['lat'])
    long=str(result['result']['geometry']['location']['lng'])
    types = result['result']['types'][0]
    results =gmaps.places_nearby(location=(lat,long),type=types,rank_by=rank, radius=500,name='Los Trompos Circuito')
    return results
    
```
