```python
#Importing libraries
import pandas as pd
import json
import requests
```

```python
def list_crime():
    #Url to work with the api
    url = 'https://hoyodecrimen.com/api/v1/crimes'
    #Getting the Url
    res = requests.get(url)
    #Converting as a json
    result = json.loads(res.content)
    return result
```

* Create a function to obtain all the crimes within a certain distance given the geographic coordinates of a point (lat/lon). Create a table locating the crimes near to the Zocalo (19.433735, -99.133334):
    * crime
    * date
    * hour
    * lat
    * lon


```python
def  crimes_longlat(crime, lat, long, distance):
    url = 'https://hoyodecrimen.com/api/v1/latlong/crimes/{}/coords/{}/{}/distance/{}'.format(crime, lat, long, distance)
    res = requests.get(url)
    result = json.loads(res.content)
    return result
```

```python
crime= 'all'
lat = '-99.133334'
long='19.433735'
distance = 5000 
zocalo = crimes_longlat(crime, lat, long, distance)
```

```python
zocaloframe = pd.read_json(json.dumps(zocalo['rows']))
```

```python
zocaloframe = zocaloframe[['crime', 'date', 'hour','lat', 'long']]
zocaloframe
```

* Obtain the top ranked quadrants the highest crime counts in 2016, 2017, 2018. Create a unique table including:
    * year
    * crime
    * quadrant
    * difference
    * population
    * rank


```python
def top_rank(crimes, start_date, end_date):
    url = 'https://hoyodecrimen.com/api/v1/cuadrantes/crimes/{}/top/counts?start_date={}&end_date={}'.format(crimes, start_date, end_date)
    res = requests.get(url)
    result = json.loads(res.content)
    return result
```

```python
topcrime1 = top_rank('all', '2016-01', '2016-12')
topcrime2 = top_rank('all', '2017-01', '2017-12')
topcrime3 = top_rank('all', '2018-01', '2018-12')
```

```python
df1= pd.read_json(json.dumps(topcrime1['rows'])) 
df2= pd.read_json(json.dumps(topcrime2['rows']))
df3= pd.read_json(json.dumps(topcrime3['rows'])) 
```

```python
main_df = pd.DataFrame()
main_df = main_df.append(df1)
main_df = main_df.append(df2)
main_df = main_df.append(df3)
```

```python
main_df['YEAR'] = main_df['start_period'].str.slice(start=0, stop=4)
```

```python
main_df = main_df.drop(columns=['count', 'end_period', 'start_period', 'sector'])
main_df = main_df[['YEAR', 'crime', 'cuadrante', 'population', 'rank']]
main_df
```
